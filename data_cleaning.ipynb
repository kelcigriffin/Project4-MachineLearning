```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
# Read the data into dataframes
df_joined = pd.read_csv('Resources/df_joined_data.csv')
df_joined.head()
```

```python
df_joined = df_joined.drop(columns=['Design Plan', 'Cathedral Ceiling', 'chicago_community_area_num', 'chicago_community_area_name', 'qualified_opportunity_zone_num'])
```

```python
#deleting Sale Date, bc Sale Year is available. Deleting Condition, Desirability and Utility, bc they are not needed for the analysis/minimal data available
df_joined = df_joined.drop(columns=['Deed No.', 'Modeling Group', 'Sale Date', 'Condition, Desirability and Utility', 'Condo Class Factor'])
```

```python
df_joined = df_joined.drop(columns=['Condo Strata', 'Sale Quarter',	'Sale Half-Year', 'Sale Quarter of Year', 'Sale Month of Year',	'Sale Half of Year', 'Most Recent Sale', 'Age Squared',	'Age Decade', 'Age Decade Squared'])
```

```python
df_joined = df_joined.drop(columns=['Pure Market Filter', 'Square root of lot size', 'Square root of age',	'Square root of improvement size'])
```

```python
#Removed Renovation, bc only one row had data
df_joined = df_joined.drop(columns=['subdivision_id', 'num_pin_in_half_mile', 'lake_michigan_data_year', 'Wall Material', 'Roof Material', 'Renovation', 'Site Desirability'])
```

```python
#Not much data in Total Building Square Feet 
df_joined = df_joined.drop(columns=['Number of Commercial Units', 'Total Building Square Feet', 'Attic Finish', 'Construction Quality'])
```

```python
#redundant columns
df_joined = df_joined.drop(columns=['class', 'tax_district_code', 'tax_district_code', 'property_address', 'flood_fs_factor', 'flood_fs_risk_direction', 'ohare_noise_contour_no_buffer_bool', 'ohare_noise_contour_half_mile_buffer_bool',	'airport_noise_dnl', 'airport_noise_data_year', 'nearest_major_road_dist_ft', 'township_code', 'property_address'])
```
