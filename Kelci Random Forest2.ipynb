## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
df_analysis = pd.read_csv("Resources/data_cleaned.csv", index_col=0)

pd.set_option('display.max_columns', None)
# Review the DataFrame
df_analysis
```

```python
# Remove columns not necessary for analysis
df_analysis = df_analysis.drop(columns=['Town and Neighborhood', 'pin', 'Property Address', 'Sale Year', 
                                        'property_city', 'property_state'])
```

```python
# Drop any additional columns here
df_analysis = df_analysis.drop(columns=['Property Class', 'Neighborhood Code', 'Land Square Feet', 'Rooms', 'Town Code', 'Type of Residence', 'Apartments', 'Use', 'Road Proximity', 
                                        'Lot Size Squared', 'Improvement Size Squared', 'longitude', 'latitude', 'property_zip', 'ward_num', 'school_elementary_district_name',
                                        'school_secondary_district_name', 'num_bus_stop_in_half_mile', 'num_foreclosure_in_half_mile_past_5_years', 'num_foreclosure_per_1000_pin_past_5_years',
                                        'nearest_cta_route_name', 'nearest_neighbor_1_dist_ft', 'nearest_neighbor_2_dist_ft', 'nearest_neighbor_3_dist_ft', 'township_name'])

df_analysis
```

```python
# drop a few more columns
df_analysis = df_analysis.drop(columns = ['Attic Type','Garage 1 Size', 'Garage 1 Attachment', 'Garage 1 Area', 'Age',
                                          'cmap_walkability_no_transit_score',	'cmap_walkability_total_score',	
                                          'nearest_cta_stop_dist_ft',	'lake_michigan_dist_ft',	'nearest_metra_stop_dist_ft',	
                                          'nearest_water_dist_ft' ])
```

```python
# Check and remove null values
nan_count = df_analysis.isnull().sum().sum()
print('Number of NaN values:', nan_count)
df_analysis = df_analysis.dropna()
df_analysis
```

```python
# Get rid of outlier sale prices
df_analysis = df_analysis.drop(df_analysis[df_analysis['Sale Price'] <= 1000].index)
#df_analysis = df_analysis.drop(df_analysis[df_analysis['Sale Price'] == 0].index)
pd.set_option('display.max_columns', None)
df_analysis
```

```python
df_analysis.dtypes
```

```python
df_analysis = df_analysis.drop(columns=['Basement Finish', 'Central Heating', 'Other Heating'])
```

### Step 2: Create the labels set (`y`)  from the â€œloan_statusâ€ column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features

# Separate the y variable, the labels
# YOUR CODE HERE!]
y = df_analysis["Sale Price"]

# Separate the X variable, the features
X = df_analysis.drop(columns=["Sale Price"])
```

```python
# Review the y variable Series
y.head()
```

```python
# Review the X variable DataFrame
X.head()
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state to the function
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=42)
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
```

```python
# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)
```

```python
# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
predictions = rf_model.predict(X_test_scaled)
```

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
results.head(10)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

### Step 3: Evaluate the modelâ€™s performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
# Generate a confusion matrix for the model
cm = confusion_matrix(y_test, predictions)
```

```python
# Print the classification report for the model
print(classification_report(y_test, predictions))
```
